```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('models/EfficientNetV2S/EfficientNetV2S.log', sep=',', header='infer')
```

```python
df.head()
```

```python
Acc_diff = df['val_accuracy'] - df['accuracy']
Loss_diff= df['val_loss'] - df['loss']
```

```python
# Add column to dataframe
df['Acc_diff'] = Acc_diff
df['Loss_diff'] = Loss_diff
```

```python
# df.head()
df
```

```python
plt.plot(df['epoch'], Acc_diff, label='Acc_diff')
plt.plot(df['epoch'], Loss_diff, label='Loss_diff')
plt.title('Accuracy and Loss difference')
plt.legend()
plt.show()
```

```python
# scatter plot of accuracy vs val_accuracy
plt.figure(figsize=(18, 6))
plt.subplot(1,3,1)
plt.plot(df['epoch'], df['accuracy'], label='accuracy')
plt.plot(df['epoch'], df['val_accuracy'], label='val_accuracy')
plt.plot(df['epoch'], df['loss'], label='loss')
plt.plot(df['epoch'], df['val_loss'], label='val_loss')
plt.title('Accuracy vs Validation Accuracy & Loss vs Validation Loss')
plt.legend()

plt.subplot(1,3,2)
plt.plot(df['epoch'], df['accuracy'], label='accuracy')
plt.plot(df['epoch'], df['val_accuracy'], label='val_accuracy')
plt.title('Accuracy vs Validation Accuracy')
plt.legend()

plt.subplot(1,3,3)
plt.plot(df['epoch'], df['loss'], label='loss')
plt.plot(df['epoch'], df['val_loss'], label='val_loss')
plt.title('Loss vs Validation Loss')
plt.legend()

plt.show()
```
